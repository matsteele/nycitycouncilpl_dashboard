```python
import pickle 
from bs4 import BeautifulSoup
import selenium
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import pandas as pd
options = webdriver.ChromeOptions()
driver = webdriver.Chrome(chrome_options=options)
```

```python
soup = BeautifulSoup(driver.page_source, 'html.parser')
content = soup.find('section', class_='post-content')
date = soup.find('p', class_='byline')
title = soup.find('h2', class_='header-xlarge')
content.text
```

```python
def scrape():
    titleArr = []
    dateArr = []
    contentArr = []
    linksArr = []
    
    for i in range(10,1942):
        driver.get('https://council.nyc.gov/press/2014/09/21/{}/'.format(i))
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        
        if soup.find('h2', class_='header-xlarge'):
            title = soup.find('h2', class_='header-xlarge').text
            date = soup.find('p', class_='byline').text
            content = soup.find('section', class_='post-content').text

            titleArr.append(title)
            dateArr.append(date)
            linksArr.append('https://council.nyc.gov/press/2014/09/21/{}/'.format(i))

            content.replace('\n','')
            content.replace('#','')

            contentArr.append(content)   
    
    return pd.DataFrame({ 'date': dateArr, 'title':titleArr, 'content':contentArr, 'link':linksArr  })
        

df = scrape()
```

```python
print(df.shape)
d2 = df[df['content'].str.len() > 60]
d2.shape

pickle.dump(d2, open('200421_press_releases.pickle', "wb" ))

## load all dataframes
# df2 = pickle.load( open( 'press_releases.pickle', "rb" ))
# df2.iloc[1508]['content']
```
